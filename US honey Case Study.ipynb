# Q3.What is the Change in mean Average price of Honey 

```python
chg_avg_price = df.groupby('year')['average_price'].mean()
chg_avg_price
```

```python
new_avg_price = chg_avg_price = df.groupby('year')['average_price'].mean().reset_index()
new_avg_price
```

 # Which was the year when production of Honey in whole US was the highest?

```python
df.groupby("year")["production"].sum()
```

```python
highest_prdct = df.groupby("year")["production"].sum().sort_values(ascending=False).reset_index()
highest_prdct.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=highest_prdct.year,y=highest_prdct.production)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#the honey production was highest in year 200 in the entire period of production
```

# Q5.from the above inference we get the production was highest in the year 2000, now let infer which state was having highest contribution in that year?

```python
max_state_2000=df[df['year']==2000]
max_state_2000
```

```python
max_state_2000 = max_state_2000.sort_values('production',ascending=False)
max_state_2000
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=max_state_2000,x='state',y='production')
plt.xticks(rotation=90)
plt.show()
```

```python
# northDakota in 2000 was the state which have highest no of honey production in 2000
```
